```python
import random
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
```

```python
#To show how we can sample from a Poisson distribution by leveraging the fact that the time between events is
#exponentially distributed, we will simply sample events one at a time, accumulating time until we hit our target
#time.
```

```python
#function to sample from poisson 1 time
def poisson(target_time, rate):
    i = 0
    events = -1
    #we simply accumulate exponentially sampled amounts of time
    while i < target_time:
        i += np.random.exponential(1 / rate)
        events += 1 #after each sample we increment number of events
    return events
```

```python
#function to sample from poisson n times
def n_poisson(n, target_time, rate):
    event_table = np.array([])
    for iterations in range(n):
        event_table = np.append(event_table, poisson(target_time, rate))
    return event_table
```

```python
event_table = n_poisson(1000, 1, 1)
```

```python
#for lambda = 1
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=False)
plt.title("Density of Poisson from my algorithm vs NumPy's algorithm", loc='right')
plot = axs[0].hist(event_table, density=True)
plot = axs[1].hist(np.random.poisson(lam=rate, size=1000), density=True, color='r')
```
